# Question 3 a)

```python
continuer = True
while continuer is True:
    try:
        val = int(input("""Which operation would you like to perform next?
1. Add (+)
2. Subtract (-)
3. Multiply (*)
4. Divide (normally) (/)
5. Integer Division (//)
6. Exponentiation (**)

Enter your choice: """))
    except ValueError:
        print("Please enter a valid int between 1 and 6: ")
        continuer = True
        continue
    if 1 <= val <= 6:
        x = float(input("Enter first value: "))
        y = float(input("Enter second value: "))
        if val == 1:
            print("x+y =",(x+y))
        elif val == 2:
            print("x-y =",(x-y))
        elif val == 3:
            print("x*y =",(x*y))
        elif val == 4:
            print("x/y =",(x/y))
        elif val == 5:
            print("x//y =",(x//y))
        elif val == 6:
            print("x**y =",(x**y))
    else:
        print("Please enter a valid number (1, 2, 3, 4, 5 or 6)")
        continuer = True
        continue
        
    inputNotOk = True
    while inputNotOk:
        t = input("Do you want to perform another calculation? (Enter y/n): ")
        if t=='y':
            inputNotOk = False
            print("Starting next round...")
        elif t=='n':
            continuer = False
            inputNotOk = False
            print("Exiting the loop")
        else:
            inputNotOk = True
            print("Please enter a valid option...")
```

# Question 3 b)

```python
continuer1 = True
while continuer1:
    try:
        n = int(input("Enter the number of items you want to buy: "))
        continuer1 = False
    except ValueError:
        print("Please enter a valid number...")
        continuer1 = True

continuer = True

items = []
costs = []
qntys = []

print(items)

i = 0
while i < n:
    item = input("Enter an Item name: ")
    cost = float(input("Enter the Cost: "))
    qty = int(input("Enter the Quantity you wish to purchase: "))
    
    if len(item) > 14:
        item = item[:14]
    elif len(item) < 14:
        item = item + " "*(14-len(item))
    if len(str(cost)) > 7 and len(str(qty)) > 7 or cost==0.0:
        print("Please enter a more realistic price and quantity ;)")
        continue
    elif len(str(cost)) > 4 or cost==0.0:
        print("Please enter a more realistic price ;)")
        continue
    elif len(str(qty)) > 4 or qty == 0:
        print("Please enter a more realistic (and non-zero) quantity ;)")
        continue
    
    items.append(item)
    costs.append(cost)
    qntys.append(qty)
    i+=1

print("\n")
print("\t\t\t\tPlus Point Hypermarket\n")
print("| Name           | Cost \t\t| Quantity \t| Total          |")
print(" ________________________________________________________________________________")
for j in range(n):
    s = str(costs[j]*qntys[j])
    tot = s + " "*(14 - len(s))
    print("|",items[j],"|",costs[j],"\t\t|",qntys[j],"\t\t|",tot,"|")

print(items)
```
