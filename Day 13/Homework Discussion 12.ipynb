```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

maxplayers = 5
maxmoney = 300
numgames = 1000
lessthan7 = []
equals7 = []
morethan7 = []
numplayers = np.random.randint(1,maxplayers+1,numgames)

for i in range(numgames):
    lessthan7.append(np.random.randint(0,numplayers[i]+1))
    equals7.append(np.random.randint(0,numplayers[i]-lessthan7[i]+1))
    morethan7.append(numplayers[i]-lessthan7[i]-equals7[i])

die1 = np.random.randint(1,7,numgames)
die2 = np.random.randint(1,7,numgames)
sumdice = die1 + die2
moneyrcvd = maxmoney * numplayers
payout = []

for i in range(numgames):
    if sumdice[i] > 7:
        payout.append(morethan7[i]*20)
    elif sumdice[i] < 7:
        payout.append(lessthan7[i]*20)
    elif sumdice[i] == 7:
        payout.append(equals7[i]*30)

retained = moneyrcvd - payout
cumul = []
summation = 0
for i in range(numgames):
    summation+=int(retained[i])
    cumul.append(summation)

plt.plot(cumul)
plt.xlabel("Number of Games")
plt.ylabel("Cumulative Money Retained")
plt.title("Cumulative Amount")
plt.show()
```

```python
table = pd.DataFrame({
    "No. of Players": numplayers,
    "Less than 7": lessthan7,
    "Equal to 7": equals7,
    "More than 7": morethan7,
    "Die 1": die1,
    "Die 2": die2,
    "Total Points": sumdice,
    "Game Money Received": moneyrcvd,
    "Payout": payout,
    "Retained": retained,
    "Cumulative Money Retained": cumul
})
pd.set_option("max_rows",100)
table
```
