## Fancy Printing in Python

### Format Strings

In C, it was possible to use a format string. This is also possible in Python (v3.6 onwards)

````python
name = input("Enter your name: ")
number_of_apples = 9
some_text = "Hello %s I hope you are well! Apparently you have %d apples!" % (name, number_of_apples)
print(some_text)
````

This is another way of achieving the same thing:

````python
name = input("Enter your name: ")
number_of_apples = 9
some_text = f"Hello {name} I hope you are well! Apparently you have {number_of_apples} apples!"
print(some_text)
````

And, the way we've seen before:

````python
name = input("Enter your name: ")
number_of_apples = 9
print("Hello", name, "I hope you are well! Apparently you have", number_of_apples, "apples!")
````
