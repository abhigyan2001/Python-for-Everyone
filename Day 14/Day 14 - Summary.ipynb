### 1. Gradient

A note on the Gradient which we saw yesterday. It'll be useful later, so definitely learn it properly:
https://en.wikipedia.org/wiki/Gradient

Also, Gradient points in the direction of **steepest ASCENT** not descent. Learn more about it here: https://www.khanacademy.org/math/multivariable-calculus/multivariable-derivatives/partial-derivative-and-gradient-articles/a/the-gradient

```python
# This code is what was shown towards the end yesterday, from my college Jupyter Notebooks:

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.4)
y = x
def gauss(x, y, x0, y0, sd):
    r = (x-x0)**2 + (y-y0)**2;
    return np.exp(-(r/sd)**2)
z = np.zeros([25, 25])
for i in range(0,25,1):
    for j in range(0,25,1):
        z[i,j] = gauss(x[i], y[j], 2.5, 2.5, 4) - gauss(x[i], y[j], 7.5, 7.5, 2)

gx = np.zeros([24, 24]);
gy = np.zeros([24, 24]);

for i in range(2,24,1):
    for j in range(2,24,1):
        #Note: i represents the rows, while j represents the columns. Hence, i = y-axis, and j = x-axis.
        gx[i,j] = -1*(z[i,j+1] - z[i, j-1])/2
        gy[i,j] = -1*(z[i+1,j] - z[i-1, j])/2

plt.contourf(z, 50, cmap="hot")
plt.colorbar()
plt.quiver(gx,gy)
plt.xlabel('Distance along x')
plt.ylabel('Distance along y')
plt.title('A hill and a valley')
plt.text(2,14,"The arrows show the direction of the -ve gradient", fontsize=10)
plt.show()
```

### 2. Sigmoid Function

$\text{sigmoid}(t) = \frac{1}{1+e^{-t}}$

It was used as an activation function in neural networks for a really long time. Now, it has been replaced by functions that produce much more accurate results, but whenever someone is taught about Neural Networks for the first time, the Sigmoid function is first used to introduce the topic, and then the other state-of-the-art functions are taught so that the learner gets a very clear idea of why this function needed to be replaced with another one.

```python
x = np.arange(-5,5,0.1)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(x,1/(1+np.exp(-1*x)))
plt.title(r"Sigmoid Function: sigmoid$(t) = \frac{1}{1+e^{-t}}$")
```
