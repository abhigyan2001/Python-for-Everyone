  ```python
  def matchingTheString(pattern, test):
      if pattern[0]=='^' and pattern[1:] == test[0:len(pattern)-1]:
          return True
      elif pattern[-1]=='$' and pattern[0:len(pattern)-1] == test[len(test)-len(pattern)+1:]:
          return True
      elif pattern in test:
          return True
      else:
          return False
  ```
